```python
import os
from glob import glob 
import re
import pandas as pd
import pathlib
from shutil import copyfile
from datetime import date
```

```python
sep = os.path.sep
usr = os.getlogin()
box_dir = sep + 'Users' + sep + usr + sep + 'Box'+ sep + 'SeroNet Public Data'
```

```python
today = date.today()
today = today.strftime("%y-%m-%d")

basic_template = '/Users/liualg/Documents/SeroNet/template/basic_study_design.xlsx'
protocol_template = '/Users/liualg/Documents/SeroNet/template/protocols.xlsx'
df = pd.read_csv(f'/Users/liualg/Documents/SeroNet/auto_add_masterlist/new_pmids/data_pull_{today}.csv')
df.head()
```

```python
# automatically creating file directory in box 
for i in range(len(df['Base_Project.1'])):
    try: 
        matching = [s for s in glob(box_dir + sep + '*') if df['Base_Project.1'][i] in s]
        print(matching[0].split('/')[-1])
        if os.path.exists(matching[0] + sep + 'PMID_' + str(df.PMID[i])):
            print(f'{df.PMID[i]} folder already exists\n')
        else:
            user = sep + 'Users' + sep + usr
            pathlib.Path(f"{matching[0]}/PMID_{df.PMID[i]}/submitted_data/").mkdir(parents=True, exist_ok=True)
            pathlib.Path(f"{matching[0]}/PMID_{df.PMID[i]}/templated_data/archived/").mkdir(parents=True, exist_ok=True)
            # Copying the basic design template into each folder
            copyfile(basic_template,
                     f"{matching[0]}/PMID_{df.PMID[i]}/templated_data/basic_study_template.xlsx")
            copyfile(protocol_template,
                     f"{matching[0]}/PMID_{df.PMID[i]}/templated_data/protocol.xlsx")
            
    except:
        print(f"\nBase Project does not exist for: PMID_{df.PMID[i]}\n")
        
```
