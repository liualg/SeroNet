```python
import pandas as pd
import numpy as np
from os import path 
```

```python
raw_files='/Users/liualg/Documents/GitHub/SeroNet/outbreak/_output-fast'
new_york=pd.read_csv(path.join(raw_files,'New York_data-pangoterms.csv'))
arizona=pd.read_csv(path.join(raw_files,'Arizona_data-pangoterms.csv'))
minnesota=pd.read_csv(path.join(raw_files,'Minnesota_data-pangoterms.csv'))
```

```python
ny_subset = new_york[(new_york['date'] >= '2020-08-01') & 
         (new_york['date'] < '2021-05-01') &
         (new_york['prevalence'] > 0)
        ]

az_subset = arizona[(arizona['date'] >= '2020-08-01') & 
         (arizona['date'] < '2021-05-01') &
         (arizona['prevalence'] > 0)
        ]

mn_subset = minnesota[(minnesota['date'] >= '2020-08-01') & 
         (minnesota['date'] < '2021-05-01') &
         (minnesota['prevalence'] > 0)
        ]

total = list(ny_subset['term']) + list(az_subset['term']) + list(mn_subset['term']) 
set(total)
```
